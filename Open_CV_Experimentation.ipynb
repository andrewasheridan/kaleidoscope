```python
import glob
import cv2
import random
import matplotlib.pyplot as plt
```

```python
image_dir = '../DataSets/Cats_and_Dogs/*'

image_filenames = glob.glob(image_dir)
```

```python
random_image_filename = random.choice(image_filenames)

img = cv2.imread(random_image_filename)
plt.imshow(img[...,::-1])
plt.title(random_image_filename)
plt.show()
```

```python
height, width = img.shape[:2]
```

```python
%%timeit
res = cv2.resize(img, (10 * width, 10 * height), interpolation = cv2.INTER_AREA)
```

```python
%%timeit
res = cv2.resize(img, (10 * width, 10 * height), interpolation = cv2.INTER_LANCZOS4)
```

```python
plt.imshow(res, aspect='equal')
```

```python
rotation_matrix = cv2.getRotationMatrix2D((height/2, width/2),90,1)
```

```python
dst = cv2.warpAffine(img, rotation_matrix, (height, width))
```

```python
plt.imshow(dst)
```
